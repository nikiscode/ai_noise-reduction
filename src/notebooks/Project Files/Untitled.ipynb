```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a clean optical signal (sine wave)
x = np.linspace(0, 10, 1000)
clean_signal = np.sin(2 * np.pi * 1 * x)  # 1 Hz sine wave

# Add random noise
noise = np.random.normal(0, 0.3, clean_signal.shape)
noisy_signal = clean_signal + noise

# Plot the signals
plt.figure(figsize=(12,4))
plt.plot(x, clean_signal, label='Clean Signal')
plt.plot(x, noisy_signal, label='Noisy Signal', alpha=0.6)
plt.legend()
plt.show()
```

```python
def moving_average(signal, window_size):
    return np.convolve(signal, np.ones(window_size)/window_size, mode='same')

# Apply filter
denoised_signal = moving_average(noisy_signal, window_size=5)

# Plot results
plt.figure(figsize=(12,4))
plt.plot(x, clean_signal, label='Clean Signal')
plt.plot(x, noisy_signal, label='Noisy Signal', alpha=0.5)
plt.plot(x, denoised_signal, label='Denoised Signal', color='red')
plt.legend()
plt.show()
```
